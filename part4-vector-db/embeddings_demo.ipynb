```python
sentences = [
    # Cricket (4)
    "The batsman scored a century in the match",
    "The bowler took a wicket in the final over",
    "The team won the cricket match by 5 runs",
    "Fielders played an important role in saving runs",

    # Cooking (3)
    "I added spices while cooking the curry",
    "The recipe requires fresh vegetables and herbs",
    "Baking a cake needs precise measurements",

    # Cybersecurity (3)
    "Cybersecurity protects systems from digital attacks",
    "Hackers try to exploit system vulnerabilities",
    "Encryption ensures data privacy and security"
]
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm', xticklabels=False, yticklabels=False)
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 similar sentences
top_indices = similarities.argsort()[-2:][::-1]
print("Top 2 similar sentences:\n")

for idx in top_indices:
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {similarities[idx]:.4f}")
    print()
```
